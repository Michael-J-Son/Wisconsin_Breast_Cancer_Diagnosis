# Breast Cancer Diagnosis

## Introduction

Breast cancer is cancer in which breast cells grow uncontrollably. Signs of breast cancer may encompass a lump in the breast, a change in breast morphology, dimpling of the skin, fluid coming from the nipple, a newly inverted nipple, or a red or scaly patch of skin. There are multiple types of breast cancer, and the cancer type is contingent on which breast cells are converted to cancer. The most common types of breast cancer are invasive ductal carcinoma and invasive lobular carcinoma. Features of the current dataset are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They depict characteristics of the cell nuclei within the image. This project aims to utilize machine learning to improve the overall accuracy of breast cancer diagnosis. Investigating features to which malignancy are attributable would be critical to the current project.

__Key Inquiries__:

    - How is cell size related to malignancy?
    - Is the change in cell size due to malignancy or is malignancy attributable to the change in cell size?
    - What is the relationship between cell shape and malignancy?

__Attribute (Feature) Information:__

    1) ID number
    
    2) Diagnosis (M = malignant, B = benign)
    
    3-32) Ten real-valued features are computed for each cell nucleus:
        a) radius (mean of distances from center to points on the perimeter)
        b) texture (standard deviation of gray-scale values)
        c) perimeter
        d) area
        e) smoothness (local variation in radius lengths)
        f) compactness (perimeter^2 / area - 1.0)
        g) concavity (severity of concave portions of the contour)
        h) concave points (number of concave portions of the contour)
        i) symmetry
        j) fractal dimension ("coastline approximation" - 1)
